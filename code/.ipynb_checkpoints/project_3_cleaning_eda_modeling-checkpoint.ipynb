```python
# Load the English spaCy model
nlp = spacy.load('en_core_web_sm')

# Add the language detection pipeline to the spaCy pipeline
nlp.add_pipe("language_detector")
```

```python
# Function to filter out non-English text
def filter_english_text(text):
    """
    Filters out non-English text by checking if the input text is in English.
    :param text: Input text to be filtered.
    :return: True if the text is in English, False otherwise.
    """
    # Process the input text using spaCy
    doc = nlp(text)
    # Check if the detected language is English
    if doc._.language == 'en':
        return True
    else:
        return False
```

```python
# Apply the filter_english_text function to df
df['Is_English'] = df['Comment'].apply(filter_english_text)

# Filter df to keep only rows with English text
df = df[df['Is_English']]

# Display the filtered DataFrame
df
```

```python
# Drop Is_English column as it is no longer needed
df.drop('Is_English', axis=1, inplace=True)
```

```python
# Utility function to remove html tags
def remove_html(text):
    """
    Removes HTML tags from the input text.
    :param text: Input text containing HTML tags.
    :return: Text with HTML tags removed.
    """
    # Define a regular expression pattern to match HTML tags
    html = re.compile(r"<.*?>")
    # Use the regular expression to remove HTML tags from the text
    return html.sub(r"", text)

# Utility function to remove urls
def remove_url(text):
    """
    Removes URLs (web links) from the input text.
    :param text: Input text containing URLs.
    :return: Text with URLs removed.
    """
    # Define a regular expression pattern to match URLs
    url = re.compile(r"https?://\S+|www\.\S+")
    # Use the regular expression to remove URLs from the text
    return url.sub(r"", text)

# Utility function to remove spaces and tabs
def remove_spaces_tabs(text):
    """
    Removes extra spaces and tabs from the input text.
    :param text: Input text containing spaces and tabs.
    :return: Text with extra spaces and tabs removed.
    """
    # Split the text into words and then join them with a single space
    return " ".join(text.split())

# Utility function to remove single character
def remove_single_char(text):
    """
    Removes standalone single characters from the input text.
    :param text: Input text containing single characters.
    :return: Text with standalone single characters removed.
    """
    # Define a regular expression pattern to match standalone single characters
    return re.sub(r"\b[a-zA-Z]\b", "", text)

# Utility function to expand contractions
def expand_contractions(text):
    """
    Expands contractions in the input text.
    :param text: Input text containing contractions.
    :return: Text with contractions expanded.
    """
    # Use the contractions library to fix contractions in the text
    return contractions.fix(text)
```

```python
# Apply the remove_html function to df
df["clean"] = df["Comment"].map(lambda x: remove_html(x))

# Apply the remove_url function to df
df["clean"] = df["clean"].map(lambda x: remove_url(x))

# Apply the remove_spaces_tabs function to df
df["clean"] = df["clean"].map(lambda x: remove_single_char(x))

# Apply the remove_single_char function to df
df["clean"] = df["clean"].map(lambda x: expand_contractions(x))

# Apply the expand_contractions function to df
df["clean"] = df["clean"].map(lambda x: remove_spaces_tabs(x))
```

```python
# Drop the 'Comment' column as we will now work with the 'clean' column
df.drop('Comment', axis=1, inplace=True)
```

```python
# Replace empty strings with Na value
df['clean'].replace('', np.nan, inplace=True)
```

[A zero-shot sentiment analysis model uses zero-shot learning (ZSL) which refers to building a model and using it to make predictions on tasks the model was not trained to do.](https://medium.com/@AmyGrabNGoInfo/sentiment-analysis-hugging-face-zero-shot-model-vs-flair-pre-trained-model-57047452225d) They can be used on any text classification task, including but not limited to sentiment analysis and topic modeling.     - The third method uses a lexicon-based sentiment analysis library. [The lexicon approach has a mapping between words and sentiment, and the sentiment of a sentence is the aggregation of the sentiment of each term.](https://medium.com/@chyun55555/unsupervised-sentiment-analyis-with-sentiwordnet-and-vader-in-python-a519660198be) I use the VADER library which is focused on social media and puts a lot of effort into identifying the sentiments of content that typically appear on social media, such as emojis, repetitive words, and punctuations (exclamation marks, for example)

## 7.2 Sentiment Analysis using Flair pre-trained sentiment model

```python
# Create our flair classifier for sentiment analysis
flair_classifier = TextClassifier.load('en-sentiment')
```

```python
# Define a function to get Flair sentiment prediction score
def score_flair(text):
    """
    Predicts sentiment using Flair sentiment analysis and returns the predicted label.
    
    :param text: Input text to be analyzed.
    :return: Predicted sentiment label.
    """  
    # Flair tokenization
    sentence = Sentence(text)
    # Predict sentiment
    flair_classifier.predict(sentence)
    # Extract the score of the predicted sentiment label
    score = sentence.labels[0].score
    # Extract the predicted sentiment label
    value = sentence.labels[0].value
    # Return the score and the predicted label
    return score, value
```

```python
# Create an empty dataframe to store the results
flair_df = pd.DataFrame()
```

```python
# Get sentiment score for each review
flair_df['scores_flair'] = mini_df['clean'].apply(lambda s: score_flair(s)[0])

# Predict sentiment label for each review
flair_df['pred_flair'] = mini_df['clean'].apply(lambda s: score_flair(s)[1])

# Change the label of flair prediction to 0 if negative and 1 if positive
mapping = {'NEGATIVE': 0, 'POSITIVE': 1}
flair_df['pred_flair'] = flair_df['pred_flair'].map(mapping)

# Assign the text column to flair_df
flair_df['clean'] = mini_df['clean']
```

## 7.3 Sentiment Analysis using Vader lexicon-based sentiment analysis library

```python
# Create a SentimentIntensityAnalyzer object
sia = SentimentIntensityAnalyzer()

# Function to classify sentiment labels
def get_sentiment_label(text):
    """
    Classifies the sentiment label of the given text based on sentiment scores.
    
    :param text: Input text to be classified.
    :return: Sentiment label ('Positive' or 'Negative').
    """
    # Calculate sentiment scores using VADER sentiment analysis
    sentiment_scores = sia.polarity_scores(text)
    # Determine sentiment label based on the compound score
    if sentiment_scores['compound'] >= 0.00:
        return 'Positive'
    elif sentiment_scores['compound'] < 0.00:
        return 'Negative'
```

```python
# Create an empty dataframe to store the results
vader_df = pd.DataFrame()
```

```python
# Apply sentiment analysis to the DataFrame column
vader_df['sentiment_label'] = mini_df['clean'].apply(get_sentiment_label)

# Change the label of vader prediction to 0 if negative and 1 if positive
vader_mapping = {'Negative': 0, 'Positive': 1}
vader_df['sentiment_label'] = vader_df['sentiment_label'].map(vader_mapping)

# Recast 'sentiment_label' datatype as int
vader_df['sentiment_label'] = vader_df['sentiment_label'].astype('int')

# Assign the text column to our vader_df
vader_df['clean'] = mini_df['clean']
```

## 7.4 Sentiment Analysis Results for all 3 Methods

```python
def calculate_specificity_npv_harmonic(y_true, y_pred):
    """
    Calculate specificity, negative predictive value (NPV), and their harmonic mean for binary classification.

    :param y_true: The true labels.
    :param y_pred: The predicted labels.
    :return: A tuple containing specificity, NPV, and their harmonic mean.
    """
    # True positives (TP): Number of correctly predicted positive samples
    TP = sum((y_true == 1) & (y_pred == 1))

    # True negatives (TN): Number of correctly predicted negative samples
    TN = sum((y_true == 0) & (y_pred == 0))

    # False negatives (FN): Number of positive samples incorrectly predicted as negative
    FN = sum((y_true == 1) & (y_pred == 0))

    # False positives (FP): Number of negative samples incorrectly predicted as positive
    FP = sum((y_true == 0) & (y_pred == 1))

    # Specificity (True Negative Rate): TN / (TN + FP)
    specificity = TN / (TN + FP) if (TN + FP) > 0 else 0.0

    # Negative Predictive Value (NPV): TN / (TN + FN)
    npv = TN / (TN + FN) if (TN + FN) > 0 else 0.0

    # Harmonic Mean of Specificity and NPV
    hm = 2 * (specificity * npv) / (specificity + npv) if (specificity + npv) > 0 else 0.0

    return specificity, npv, hm
```

```python
# Function to convert emojis to text
def convert_emojis_to_text(text):
    """
    Converts emojis in the input text to their textual representations.
    
    :param text: Input text containing emojis.
    :return: Text with emojis replaced by textual representations.
    """
    # Use emoji library to convert emojis to text
    return emoji.demojize(text)
```

```python
# Apply the convert_emojis_to_text function to the dataframe
combined_df['sequence'] = combined_df['sequence'].apply(convert_emojis_to_text)
```

```python
# Remove punctuation and non-Roman characters
def clean_text(text):
    """
    Cleans and preprocesses the input text by removing punctuation and non-Roman characters.
    
    :param text: Input text to be cleaned.
    :return: Cleaned text with punctuation and non-Roman characters removed.
    """
    # Remove punctuation using regular expression
    text = re.sub("[%s]" % re.escape(string.punctuation), "", text)    
    # Remove non-Roman characters using regular expression
    text = re.sub("([^\x00-\x7F])+", " ", text)
    return text
```

```python
# Apply the clean_text function to the dataframe
combined_df["sequence"] = combined_df["sequence"].map(lambda x: clean_text(x))
```

```python
# Utility function to convert digits to words
def convert_digits_to_words(text):
    """
    Converts numeric digits in the input text to their word representations.
    
    :param text: Input text containing numeric digits.
    :return: Text with numeric digits replaced by word representations.
    """
    # Split the text into words, convert digits to words if present
    return " ".join([num2words(w) if w.isdigit() else w for w in text.split()])

# Utility function to convert text to lower case
def lower_case(text: str) -> str:
    """
    Converts the input text to lowercase.
    
    :param text: Input text to be converted to lowercase.
    :return: Text converted to lowercase.
    """
    # Convert the input text to lowercase
    return text.lower()
```

```python
# Apply the convert_digits_to_words function to the DataFrame
combined_df["sequence"] = combined_df["sequence"].map(lambda x: convert_digits_to_words(x))

# Apply the lower_case function to the DataFrame
combined_df["sequence"] = combined_df["sequence"].map(lambda x: lower_case(x))
```

```python
# Utility function to remove stop words
def remove_stopwords(text):
    """
    Removes stopwords from the input text.
    
    :param text: Input text to be processed.
    :return: Text with stopwords removed.
    """
    # Get a set of English stopwords
    stop_words = set(stopwords.words("english"))
    # Tokenize the input text into words
    word_tokens = word_tokenize(text)
    # Filter out stopwords from the tokenized words
    filtered_sentence = [w for w in word_tokens if not w in stop_words]
    # Join the filtered words back into a single string
    filtered_text = " ".join(filtered_sentence)
    
    return filtered_text
```

```python
# Apply the remove_stopwords function to the DataFrame
combined_df['sequence'] = combined_df['sequence'].apply(remove_stopwords)
```

```python
# Initialize the SpellChecker object
spell = SpellChecker()

# Define a function to correct spelling using the SpellChecker
def correct_spelling(text):
    """
    Corrects spelling errors in the input text using SpellChecker.

    :param text: The input text containing potential spelling errors.
    :return: The corrected text with spelling errors fixed, or None if text is None.
    """
    # Check if the input text is None
    if text is None:
        return None

    # # Check if text[6] is None
    # if text[6] is None:
    #     return None

    # Tokenize the text into individual words
    tokens = text.split()

    # Correct spelling for each token using the SpellChecker
    corrected_tokens = [spell.correction(token) if spell.correction(token) is not None else token for token in tokens]

    # Join the corrected tokens back into a coherent sentence
    corrected_text = ' '.join(corrected_tokens)

    # Return the corrected text
    return corrected_text
```

```python
# Function to plot a frequency distribution of words
def plot_freq_dist(text, n=10):
    """
    Plots a frequency distribution of terms in the input text.

    :param text: A pandas Series containing text data.
    :return: None
    """
    # Initialize a CountVectorizer to convert text into a document-term matrix
    vectorizer = CountVectorizer()
    
    # Transform the text into a document-term matrix
    docs = vectorizer.fit_transform(text)
    
    # Get the feature names (terms) from the vectorizer
    features = vectorizer.get_feature_names_out()

    # Initialize a FreqDistVisualizer to visualize the frequency distribution
    visualizer = FreqDistVisualizer(features=features, n=n)
    
    # Fit the visualizer to the document-term matrix
    visualizer.fit(docs)
    
    # Display the visualization
    visualizer.poof()
```

```python
# Function to get a dataframe with n-gram sorted by frequency and create a seaborn barplot
def get_ngram_freq(n, df, frequency=10):
    """
    Calculates the frequency of n-grams in the 'sequence' column of a DataFrame and creates a seaborn barplot.

    :param n: The desired n-gram size.
    :param df: The input DataFrame containing a 'sequence' column.
    :return: A DataFrame with n-gram frequencies sorted by frequency.
    """
    # Initialize CountVectorizer with specified parameters
    vec = CountVectorizer(stop_words='english', 
                          ngram_range=(n, n))

    # Create a bag-of-words matrix of n-grams (using a sparse matrix)
    bow = vec.fit_transform(df["sequence"])

    # Sum the occurrences of each n-gram across all documents
    count_values = bow.sum(axis=0).A1

    # Create a DataFrame from n-gram frequencies
    ngram_freq = pd.DataFrame({"ngram": vec.get_feature_names_out(), "frequency": count_values})
    
    # Sort the DataFrame by frequency in descending order
    ngram_freq.sort_values(by="frequency", ascending=False, inplace=True)
    
    # Get the top 'frequency' number of n-grams
    top_ngrams = ngram_freq.head(frequency)
    
    # Create a seaborn barplot
    plt.figure(figsize=(10, 6))
    sns.barplot(x="frequency", y="ngram", data=top_ngrams, palette="viridis")
    plt.xlabel("Frequency")
    plt.ylabel("N-gram")
```

```python
# Function to retrieve most common words with a specific POS tag. Tag such as "ADJ' can be chosen and passed into the function
def get_most_common_pos_tag(df, tag):
    """
    Retrieves the most common words with a specific part-of-speech tag from a DataFrame.

    :param df: The input DataFrame containing a 'text' column.
    :param tag: The desired part-of-speech tag (e.g., 'ADJ' for adjectives).
    :return: Lists of the most common words with the specified tag and their frequencies.
    """
    # Get the set of stopwords for English
    stop = set(stopwords.words('english'))

    # Get the 'text' column values from the DataFrame
    corpus = df["sequence"].values.tolist() 
    
    # Tokenize and perform part-of-speech tagging on the entire corpus
    tokens = word_tokenize(" ".join(corpus))
    tags = nltk.pos_tag(tokens, tagset='universal')

    # Filter and store words with the specified part-of-speech tag and exclude stopwords
    tags = [word for word, pos in tags if ((pos == tag) & (word not in stop))]

    # Calculate the most common words with the specified tag and their frequencies
    most_common = Counter(tags).most_common(10)

    # Initialize lists to store most common words and their frequencies
    words, frequency = [], []

    # Iterate through the most common words and their counts
    for word, count in most_common:
        words.append(word)
        frequency.append(count)
        
    # Create a bar plot using Seaborn
    plt.figure(figsize=(10, 6))
    sns.barplot(x=frequency, y=words, palette='viridis')
    plt.xlabel('Frequency')
    plt.ylabel('Word')
```

**Steps to take part 1:**

    - This is preferable so that the negative comments are now the 'POSITIVE CLASS' and using f1-score is now appropriate. This also has the advantage of being potentially less confusing to the reader
3. Check for class inbalance. Class inbalance will be handled using class weights and class priors in-built into sklearn estimators. This has the advantage of reducing time complexity compared to using the SMOTE library to handle class inbalance. [There also exists the possibility that using class weights and class priors may improve the model's performance relative to using SMOTE or bootstrapping](https://towardsdatascience.com/stop-using-smote-to-treat-class-imbalance-take-this-intuitive-approach-instead-9cb822b8dc45)
    - Class weights are calculated using the formula (no. of samples of minority class * weight of minority class = no. of samples of majority class * weight of majority class). As such, setting the class weight of the majority class to 1, then the weight of the minority class is the ratio of the number of samples of the majotrity class to the minority class.
    - Class priors are calculated as the reciprocal of class weights, with our minority class (negative comments) having the smaller probability compared to the majority class (positive comments)
5. Create a copy of the dataframe and assign to combined_df_copy

**Steps to take part 2:**

1. **Multinomial Naive Bayes is the first model to be evaluated and will serve as our baseline model**. There are 4 combinations of preprocessing and gridsearch is used to tune the hyperparameters:
    - SnowBall Stemmer followed by CountVectorizer
    - Word Net Lemmatizer followed by Count Vectorizer
    - SnowBall Stemmer followed by TF-IDF Vectorizer
    - Word Net Lemmatizer followed by TF-IDF Vectorizer
2. The best choice of preprocessing along with the grid search selected best hyperparameters is chosen among these 4 for further evaluation of other model types
3. The next model type to be evaluated is Logistic Regression. The comparison will be between our:
    - previously chosen preprocessers and hyperparameters (this is a non-word embedding model)
    - a Word2Vec vector without stemming or lemmatizing (this is a word embedding model) meaning that a vectorized output of word embeddings will be used to train our Logistic Regression model. I have opted not to stem or lemmatize before passing the text into the Word2Vec model to minimize the number of model combinations to train on. **This may have biased the performance of the word embedding models and should be considered as a recommendation for future work.**
    - Multinomial Naive Bayes could not be used to evaluate a Word2Vec model as the vectorized output of a Word2Vec model can be negative. MultinomialNB cannot take in -ve values as an input
4. We will continue evaluating further models using the Word2Vec vectors and also for the previously chosen preprocessors and hyperparameters (non-word embedding) using the following models:
    - Random Forest Classifier

```python
#Initialize the Snowball Stemmer
stemmer = SnowballStemmer('english')
```

```python
# Function to stem words using a chosen stemmer, can be custom or ready-made
def stem_text_custom(text, stemmer):
    """
    Stems the words in the input text using a custom stemmer.

    :param text: The input text to be stemmed.
    :param stemmer: The custom stemmer object to perform stemming.
    :return: The text with stemmed words.
    """
    # Tokenize the input text into individual words
    word_tokens = word_tokenize(text)
    
    # Perform stemming on each word using the custom stemmer
    stemmed_output = " ".join([stemmer.stem(w) for w in word_tokens])
    
    # Return the text with stemmed words
    return stemmed_output
```

```python
#Initialize the Word Net Lemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
# Function to lemmatize words using a chosen lemmatizer, can be custom or ready made
def lemmatize_text_custom(text, lemmatizer):
    """
    Lemmatizes the words in the input text using a custom lemmatizer.

    :param text: The input text to be lemmatized.
    :param lemmatizer: The custom lemmatizer object to perform lemmatization.
    :return: The text with lemmatized words.
    """
    # Map Treebank POS tags to WordNet POS tags for lemmatization
    wordnet_map = {
        "J": wordnet.ADJ,
        "N": wordnet.NOUN,
        "V": wordnet.VERB,
        "R": wordnet.ADV,
    }
    # Tokenize the input text and perform part-of-speech tagging
    w_pos_tags = nltk.pos_tag(word_tokenize(text))
    
    # Lemmatize each word using the custom lemmatizer and the appropriate WordNet POS tag
    lemmatized_output = " ".join([lemmatizer.lemmatize(w, wordnet_map.get(pos[0], wordnet.NOUN)) for w, pos in w_pos_tags])
    
    # Return the text with lemmatized words
    return lemmatized_output
```

```python
def vectorize(sentence, size=200):
    """
    Vectorizes a sentence using a word2vec model.
    
    :param sentence: Input sentence to be vectorized.
    :return: A vector representation of the sentence.
    """
    # Split the sentence into individual words
    words = sentence.split()
    
    # Initialize an empty list to store word vectors
    words_vecs = [w2v_model.wv[word] for word in words if word in w2v_model.wv]
    
    # Check if there are no valid word vectors
    if len(words_vecs) == 0:
        # Return a zero vector (100-dimensional) if no valid word vectors are found
        return np.zeros(size)
    
    # Convert the list of word vectors to a NumPy array
    words_vecs = np.array(words_vecs)
    
    # Compute the mean vector of all word vectors
    mean_vector = words_vecs.mean(axis=0)
    
    # Return the mean vector as the vectorized representation of the sentence
    return mean_vector
```

```python
def grid_search_best(grid_search_object):
    """
    Retrieve the best cross-validated f1-score, best parameters, and return as a formatted message.

    :param grid_search_object: A trained GridSearchCV object.
    :return: A formatted message with the best score and best parameters.
    """
    # Get the best score and best parameters
    best_score = grid_search_object.best_score_
    best_params = grid_search_object.best_params_
    
    # Format the message with the best score and best parameters
    message = f'The best cross-validated f1-score is {np.round(best_score, 5)} and the best parameters are {best_params}'
    
    return message
```

```python
def calculate_f1_score(X_train, y_train, X_test, y_test, grid_search_object):
    """
    Calculate and print the F1 scores for the training and testing sets using a trained GridSearchCV object.

    :param X_train: Training data features.
    :param y_train: Training data labels.
    :param X_test: Testing data features.
    :param y_test: Testing data labels.
    :param grid_search_object: A trained GridSearchCV object.
    """
    # Predict labels for training and testing sets using the best estimator from grid search
    y_train_pred = grid_search_object.predict(X_train)
    y_test_pred = grid_search_object.predict(X_test)
    
    # Calculate F1 score for training set
    f1_train = f1_score(y_train, y_train_pred)

    # Calculate F1 score for testing set
    f1_test = f1_score(y_test, y_test_pred)
    
    # Print the F1 scores
    print("F1 score for the training set:", np.round(f1_train, 5))
    print("F1 score for the testing set:", np.round(f1_test, 5))
```

## 12.3 ROC AUC Curve

```python
# Create ROC AUC graph
fig = plt.subplots(figsize=(6, 6))
roc_viz = ROCAUC(best_model,
                classes=["Positive Comment", "Negative Comment"])
roc_viz.fit(X_train_vectorized, y_train_lem)
roc_viz.score(X_test_vectorized, y_test_lem)
roc_viz.poof();
```

1. Ideally, we would like the Area under Curve (AUC) to be a maximum of 1, which means that the Receiver Operating Characteristic curve extends all the way to the top left corner. The ROC is a measure of a classifier’s predictive quality that compares and visualizes the trade-off between the model’s sensitivity and specificity
2. What that means is that the classifier finds all the correct positive classes (in this case the curve is able to treat either positive or negative comments as the positive class, similar to the above classification report) without mislabelling a negative class wrongly as a positive class
3. Conversely if the classifier were extremely bad (or good, as we shall see) at doing the opposite, then the false positive rate would be a maximum of 1.0 with a true positive rate of 0.0, or at the bottom right corner. Our classifier would be able to mislabel every negative class wrongly as the positive class. This also means that a mislabelling of every positive class as the negative class. Essentially, the classifier is perfect just like with an AUC of 1, but the labels need to be swapped
4. The difference between macro and micro averaging of the AUC is that macro averaging weighs each class equally (50-50 in this case) and does not consider class inbalance whereas micro averaging weighs each sample equally (and considers class inbalance). If you have an equal number of samples for each class, then macro-average and micro-average will result in the same score. As there are more positive comments, the greater scores of the positive comments pulls the micro-average AUC upwards and we get micro-average of 0.89 vs a macro-average of 0.87

## 12.4 Precision-Recall Curve

```python
# Create precision recall curve
fig = plt.subplots(figsize=(6, 4))
prc_viz = PrecisionRecallCurve(best_model)
prc_viz.fit(X_train_vectorized, y_train_lem)
print(prc_viz.score(X_test_vectorized, y_test_lem))
prc_viz.poof();
```

1. Average precision expresses the precision-recall curve in a single number, which represents the area under the curve. It is computed as the weighted average of precision achieved at each threshold, where the weights are the differences in recall from the previous thresholds
2. Note that the only way to move along (left to right or vice versa) a precision-recall curve is to change the discrimination threshold for predicting the positive class. For the discrimination threshold chosen by our best model so as to maximise the f1-score which is 0.655 (see the above classification report), we also get a precision of 0.612 and a recall of 0.706. If one were to imagine a vertical line at a recall of 0.706, the vertical line will intersect the precision-recall curve at a horizontal value of around 0.612
3. If we look at the two extreme ends. At the highest recall of 1.0, the classifier would have 'captured' every negative comment. Sounds good right? But, at that level of recall, the precision is about 0.30, which means that  the negative comments only represent 30% of the predicted labels, with the other 70% being the false positives which are all the positive comments being mislabelled as negative comments.
4. Now the other extreme scenario would be that we don't want any mislabelling, so no false positives at all. Therefore no positive comments will be mislabelled as negative comments. Maximizing precision towards 1.0, the problem is we end up returning less and less of the negative comments. Since recall keeps decreasing as precision increases and (ignoring the weird spike at the top left of the graph) at a precision of 0.95, while there are only 5 mislabels for every 100 comments labeled as the 'positive' class. But, how many negative comments are we now getting? The recall values looks like 0.01 or 0.02 only, which means that we are only getting 1 or 2 negative comments out of every 100 comments predicted
5. Clearly there is a trade-off between precision and recall, and both cannot be maximized simultaneously. As mentioned previously, to balance this trade-off, the scoring metric will be the harmonic mean of precision and recall, or the f1-score

## 13.1 Conclusion 

The project aims to develop a machine learning classification model for predicting comments as either positive or negative on a Justin Bieber YouTube video. The dataset comprises raw comments and replies from the video, but without the target label. Therefore this requires sentiment analysis to create the target label.

Three sentiment analysis methods are employed: a pre-trained zero-shot deep-learning language model from Hugging Face, a custom sentiment analysis model named Flair trained on IMDB data, and VADER, a lexicon-based sentiment analysis library. These methods aim to classify comments as either positive or negative for downstream model classification.

The chosen hybrid approach computes the average predictions from the three methods to establish target labels, aiming to harness the strengths of each method and reduce potential labeling conflicts. Subsequently, the analysis progresses to EDA, feature engineering, selection, and modeling.

Two primary model categories undergo evaluation: non-word embedding models and word embedding models. Non-word embedding models create a corpus using either Snowball Stemmer or Word Net Lemmatizer. This corpus is then used to fit Multinomial Naive Bayes, Logistic Regression, Random Forest Classifier, Gradient Boosting Classifier, and Stacking Classifier. Word embedding models employ Word2Vec vectors before the output is used to fit Logistic Regression, Random Forest Classifier, Histogram-based Gradient Boosting Classifier and Stacking Classifier. Model hyperparameters undergo grid searches for optimal performance, with the best chosen used to fit each model.

From the evaluated models, Logistic Regression with non-word embedding (model 3A) emerges as the most proficient performer. It exhibits superior test results and reduced overfitting compared to the other models, leading to its selection as the ultimate choice for classification to address the problem statement.

## 13.2 Recommendation

**Create a Third Label ('Neutral') During Sentiment Analysis**

As mentioned during Step 7, there should rightfully be a class for comments that are neither positive or negative. This class called 'neutral' will contain comments that even to a human labeler seem ambiguous. That way, during the creation of the target labels, there won't be a bias where 'neutral' comments are classified either as positive or negative based on the library/pre-trained model. If these neutral comments are deemed to have value, we can train classifier models using one-vs-one or one-vs-rest strategies and then rank the models with the same metrics and visualizations that I have done so for my best model. Another option to handle the 'neutral' comments would be to drop that class after the label has been created and then proceed with the same binary classification strategy. If this improves the model's performance, this ties back into better addressing the problem statement which is a machine learning classification model that can predict whether a comment on a Youtube video is either positive or negative.

**Further Preprocessing of Text Before Vectorizing with a Word2Vec Model**

As mentioned during step 11, the Word2Vec model was not fitted with stemmed or lemmatized text data before the output is used later for the classification task. This may have biased the performance of the Word2Vec model. Although I have outlined in step 11.7, reasons why a word embedding model may underperform relative to a non-word embedding model. There are certain reasons why this representation of text data should lead to superior model performance, which I will outline below:

1. **Semantic Comprehension:** Word embeddings encapsulate the semantic connections between words by depicting them within a continuous vector space where akin words exist closer together. This attribute assists the model in comprehending the significance and context of words, even when they weren't encountered during training

2. **Feature Representation:** Word embeddings furnish a compact representation for words, whereas non-word embedding models frequently depend on sparse representations like one-hot encoding. Dense representations carry more information and are better at capturing the subtleties and resemblances among words

3. **Transfer Learning:** Pre-trained word embeddings can be leveraged as a form of transfer learning. Models trained on extensive datasets can internalize broad language patterns and subsequently be fine-tuned for specific tasks with other datasets such as our youtube comments dataset which should have helped to better address our problem statement. Unfortunately this did not seem to be the case, so perhaps the other factors also must be considered

4. **Word Sequence and Context:** Word embeddings encode both word sequence and context context as compared to non-word embedding models which often forfeit sequential information. The meaning and interdependencies among words within sentences or documents may improve the performance of the later classifier models

5. **Handling Out-of-Vocabulary (OOV) Instances:** Word embeddings are able to manage out-of-vocabulary words by associating them with analogous embeddings, while in contrast, non-word embedding models might grapple with unfamiliar words

Theoretically, this should improve the performance of the Word2Vec model relative to CountVectorizer or TF-IDF Vectorizer and should be explored as a recommendation to create a better classifier to better address the problem statement.
