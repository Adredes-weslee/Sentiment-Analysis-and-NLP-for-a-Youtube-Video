```python
def remove_punctuation(text):
    # Remove punctuation
    # punctuation_pattern = r'[^\w\s]'
    # text = re.sub(punctuation_pattern, '', text)
    text = re.sub("[%s]" % re.escape(string.punctuation), "", text)   
    
    return text
```

```python
df["text_clean"] = df["Comment"].map(lambda x: remove_punctuation(x))
```

```python
def clean_text(text):
    # Convert text to lowercase
    # text = text.lower()    
    # Remove non-Roman characters
    text = re.sub("([^\x00-\x7F])+", " ", text)
    
    return text
```

```python
df["text_clean1"] = df["Comment"].map(lambda x: clean_text(x))
```

```python
def remove_non_english(text):
    # Define a regular expression pattern that matches non-English characters
    non_english_pattern = re.compile('[^a-zA-Z0-9\s]')
    
    # Use the sub() function to replace non-English characters with an empty string
    cleaned_text = non_english_pattern.sub('', text)
    
    return cleaned_text
```

```python
df["text_clean2"] = df["Comment"].map(lambda x: remove_non_english(x))
```

```python
# Load the English spaCy model
nlp = spacy.load('en_core_web_sm')

# Add the language detection pipeline to the spaCy pipeline
nlp.add_pipe("language_detector")

# Function to filter out non-English text
def filter_english_text(text):
    doc = nlp(text)
    # Check if the detected language is English
    if doc._.language == 'en':
        return True
    else:
        return False

```

```python
# Apply the filter function to the DataFrame
df['Is_English'] = df['Comment'].apply(filter_english_text)
```

```python
# Filter the DataFrame to keep only rows with English text
english_df = df[df['Is_English']]

# Display the filtered DataFrame
english_df
```

```python
english_df.drop('Is_English', axis=1, inplace=True)
```

```python
def remove_html(text: str) -> str:
    html = re.compile(r"<.*?>")
    return html.sub(r"", text)
```

```python
# Apply the filter function to the DataFrame
english_df["no_html"] = english_df["Comment"].map(lambda x: remove_html(x))
```

```python
def remove_url(text: str) -> str:
    url = re.compile(r"https?://\S+|www\.\S+")
    return url.sub(r"", text)
```

```python
english_df["no_url"] = english_df["Comment"].map(lambda x: remove_url(x))
```

```python
def remove_spaces_tabs(text: str) -> str:
    return " ".join(text.split())
```

```python
english_df["no_spaces"] = english_df["no_url"].map(lambda x: remove_spaces_tabs(x))
```
